```python
import pandas as pd 
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head(2)
```

Extract additional features from the existing
columns, such as the length of the restaurant
name or address.

```python
df['len_restaurant']=df['Restaurant Name'].apply(len)
df['len_address']=df['Address'].apply(len)
df[['Restaurant Name', 'len_restaurant', 'Address', 'len_address']].head()
```

Create new features like "Has Table Booking"
or "Has Online Delivery" by encoding
categorical variables.


```python
df['Has_Table_Booking'] = df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Has_Online_Delivery'] = df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)

df[['Has Table booking', 'Has_Table_Booking', 'Has Online delivery', 'Has_Online_Delivery']].head()
```
