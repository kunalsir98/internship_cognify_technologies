```python
import pandas as pd 
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

Determine the most common price range
among all the restaurants.

```python
most_common_price_range = df['Price range'].mode()[0]
price_range_counts = df['Price range'].value_counts()

print("Most common price range:", most_common_price_range)
print("Price range distribution:")
print(price_range_counts)
```

Calculate the average rating for each price
range.

```python
# Calculate the average rating for each price range
average_rating_per_price_range = df.groupby('Price range')['Aggregate rating'].mean()

print("Average rating per price range:")
print(average_rating_per_price_range)

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
average_rating_per_price_range.plot(kind='bar', color='skyblue')
plt.title("Average Rating for Each Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.xticks(rotation=0)
plt.show()
```

```python

average_rating_per_price_range = df.groupby('Price range')['Aggregate rating'].mean()

print("Average rating for each price range:\n", average_rating_per_price_range)

highest_rated_price_range = average_rating_per_price_range.idxmax()
highest_average_rating = average_rating_per_price_range.max()


print("Price range with the highest average rating:", highest_rated_price_range)
print("Highest average rating:", highest_average_rating)
```
