```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head(2)
```

Task: Table Booking and Online Delivery

Determine the percentage of restaurants that
offer table booking and online delivery.

```python
table_booking=df['Has Table booking'].value_counts(normalize=True)*100
online_booking=df['Has Online delivery'].value_counts(normalize=True)*100

print('% of table booking')
print(table_booking)
print('% of online_booking')
print(online_booking)
```

Compare the average ratings of restaurants
with table booking and those without.

```python
average_rating_table_booking=df.groupby('Has Table booking')['Aggregate rating'].mean()
print('Average ratings of restaurnats with table booking and those without')
print(average_rating_table_booking)

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
average_rating_table_booking.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title("Average Rating for Restaurants with and without Table Booking")
plt.xlabel("Has Table Booking")
plt.ylabel("Average Rating")
plt.xticks(rotation=0)
plt.show()
```

Analyze the availability of online delivery
among restaurants with different price ranges.

```python
online_delivery_price_ranges=df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
print('Availability of online delivey among reastaurants with differnent price ranges')
print(online_delivery_price_ranges)
plt.figure(figsize=(8,8))
online_delivery_price_ranges.plot(kind='bar',stacked=True,color=['lightgreen','red'])
plt.title('Availability of online delivey among reastaurants with differnent price ranges')
plt.xlabel('Price range')
plt.ylabel('Has Online delivery')
plt.legend(title="Has Online Delivery", loc='upper right')
plt.xticks(rotation=0)
plt.show()
```
