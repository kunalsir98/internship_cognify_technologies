Create visualizations to represent the distribution
of ratings using different charts (histogram, bar
plot, etc.).

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head(3)
```

```python
plt.figure(figsize=(8,8))
sns.histplot(df['Aggregate rating'],bins=20,kde=True,color='red')
plt.title('Distribution of rating')
plt.xlabel('Aggregate rating')
plt.ylabel('Frequency')
plt.show()
```

Compare the average ratings of different cuisines
or cities using appropriate visualizations.

```python
average_rating_cuisines=df.groupby('Cuisines')['Aggregate rating'].mean().sort_index(ascending=False)
print('Average rating of different cuisines')
print(average_rating_cuisines)
plt.figure(figsize=(8,8))
average_rating_cuisines.head(10).plot(kind='bar',color='green')
plt.title('Top 10 Cuisines According to Average Rating')
plt.xlabel('Cuisines')
plt.ylabel('Aggregate rating')
plt.xticks(rotation=42)
plt.show()
```

```python
#cities
average_rating_cities=df.groupby('City')['Aggregate rating'].mean().sort_index(ascending=False)
print('Average rating of different cuisines')
print(average_rating_cities)
plt.figure(figsize=(8,8))
average_rating_cities.head(10).plot(kind='bar',color='purple')
plt.title('Top 10 cities According to Average Rating')
plt.xlabel('City')
plt.ylabel('Aggregate rating')
plt.xticks(rotation=42)
plt.show()
```

Visualize the relationship between various
features and the target variable to gain insights.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x='Aggregate rating',y='Price range',color='blue',alpha=0.7)
plt.title('Relationship between Aggregate rating and Price range')
plt.xlabel('Aggregate rating')
plt.ylabel('Price range')
plt.xticks(rotation=42)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x='Aggregate rating',y='Votes',color='yellow',alpha=0.7)
plt.title('Relationship between Aggregate rating and Votes')
plt.xlabel('Aggregate rating')
plt.ylabel('Votes')
plt.xticks(rotation=42)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x='Aggregate rating',y='Has Online delivery',color='red',alpha=0.7)
plt.title('Relationship between Aggregate rating and Has Online delivery')
plt.xlabel('Aggregate rating')
plt.ylabel('Has Online delivery')
plt.xticks(rotation=42)
plt.show()
```
