Task: Customer Preference Analysis

Analyze the relationship between the type of
cuisine and the restaurant's rating.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

```python
df.columns=df.columns.str.strip()
```

```python
df.head(2)
```

```python
cuisine_ratings=df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
print('relationship between the type ofcuisine and the restaurants rating')
print(cuisine_ratings)
plt.figure(figsize=(6,6))
cuisine_ratings.head(10).plot(kind='bar',color='skyblue')
plt.title("Top 10 Cuisines by Average Rating")
plt.xlabel('Cuisines')
plt.ylabel('Aggregate rating')
plt.xticks(rotation=42)
plt.show()
```

Identify the most popular cuisines among
customers based on the number of votes.

```python
popular_cuisines=df.groupby('Cuisines')['Votes'].mean().sort_index(ascending=False)
print('popular cuisines among customers based on the number of votes.')
print(popular_cuisines)
plt.figure(figsize=(8,6))
popular_cuisines.head(10).plot(kind='bar',color='skyblue')
plt.title('Top 10 cuisines based on votes')
plt.xlabel('Cuisines')
plt.xlabel('Votes')
plt.xticks(rotation=42)
plt.show()
```

Determine if there are any specific cuisines
that tend to receive higher ratings.

```python

plt.figure(figsize=(12, 8))
top_cuisines = df['Cuisines'].value_counts().index[:10]  
sns.boxplot(data=df[df['Cuisines'].isin(top_cuisines)], x='Cuisines', y='Aggregate rating')
plt.title("Rating Distribution by Cuisine Type")
plt.xlabel("Cuisine")
plt.ylabel("Aggregate Rating")
plt.xticks(rotation=45)
plt.show()
```
